```python
ham_laps = laps[
    (laps["Driver"] == "HAM") &
    (laps["IsAccurate"] == True) 
    ][["Driver", "LapNumber", "LapTime"]]


ham_laps["LapTimeSeconds"] = ham_laps["LapTime"].dt.total_seconds()

ham_laps["LapTimeDisplay"] = (
    ham_laps["LapTimeSeconds"]
    .apply(lambda x: f"{int(x//60):02d}:{x%60:06.3f}")
)
print(ham_laps)

print(ham_laps["LapTime"].mean())
```

```python
ham_lapsF = laps[
    (laps["Driver"] == "HAM") ][["Driver", "LapNumber", "LapTime"]]


ham_lapsF= ham_lapsF.dropna(subset=["LapTime"])
ham_lapsF["LapTimeSeconds"] = ham_lapsF["LapTime"].dt.total_seconds()

print(ham_lapsF)
ham_lapsF["LapTimeDisplay"] = (
    ham_lapsF["LapTimeSeconds"]
    .apply(lambda x: f"{int(x//60):02d}:{x%60:06.3f}")
)

print(ham_lapsF["LapTime"].mean())
```
